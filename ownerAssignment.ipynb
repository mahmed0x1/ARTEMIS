```python
from datasets import Dataset
import pyarrow.parquet as pq
import os

dataset_dir = "small_dataset_preprocessed"
bad_files = []

# Test loading each shard individually
for filename in os.listdir(dataset_dir):
    if filename.endswith(".arrow"):
        try:
            table = pq.read_table(os.path.join(dataset_dir, filename))
            print(f"✅ {filename} is valid")
        except Exception as e:
            print(f"❌ {filename} is corrupt: {str(e)}")
            bad_files.append(filename)

print(f"Corrupt files: {bad_files}")
```

```python
from datasets import load_from_disk, concatenate_datasets
import os

valid_shards = []
corrupt_dir = "small_dataset_preprocessed"

# Identify working shards
for fname in os.listdir(corrupt_dir):
    if fname.endswith(".arrow"):
        try:
            shard = load_from_disk(os.path.join(corrupt_dir, fname))
            valid_shards.append(shard)
            print(f"✅ {fname} is salvageable")
        except:
            print(f"❌ {fname} unrecoverable")

# Combine valid shards
if valid_shards:
    recovered_dataset = concatenate_datasets(valid_shards)
    print(f"Recovered {len(recovered_dataset)} examples")
else:
    print("No valid shards found - requires full rebuild")
```
